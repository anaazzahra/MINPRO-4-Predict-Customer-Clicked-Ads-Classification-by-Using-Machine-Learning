# **TASK 2: DATA CLEANING AND PRE-PROCESSING**

## **Missing Value**

```python
df.isnull().sum()
```

```python
#handling missing value
#numerical feature
for column in numericals:
  median_value = df[column].median()
  df[column].fillna(median_value, inplace = True)

#categorical feature
for column in Cats:
  mode_value = df[column].mode()[0]
  df[column].fillna(mode_value, inplace = True)
```

```python
df.isnull().sum()
```

## **Duplicated Data**

```python
df.duplicated().sum()
```

## **Extract Datetime**

```python
#Extraction Timestamp
df['Year'] = df['Timestamp'].dt.year
df['Month'] = df['Timestamp'].dt.month
df['Day'] = df['Timestamp'].dt.day

# Determine week of the month
df['Week Start'] = df['Timestamp'] - pd.to_timedelta(df['Timestamp'].dt.dayofweek, unit='d')

# Calculating week of the month
df['Week'] = df['Week Start'].dt.day // 7 + 1

# Delete temporary column 'Week Start'
df.drop(columns=['Week Start'], inplace=True)
```

```python
df.sample(3)
```

## **Feature Encoding**

```python
df.info()
```

```python
#Encoding with 'pd.get_dummies'
df_encode = df.copy()

#Encoding
gender_dummies = pd.get_dummies(df_encode['Gender'], prefix='Gender', prefix_sep='_')
city_dummies = pd.get_dummies(df_encode['city'], prefix='city', prefix_sep='_')
province_dummies = pd.get_dummies(df_encode['province'], prefix='province', prefix_sep='_')
category_dummies = pd.get_dummies(df_encode['category'], prefix='category', prefix_sep='_')
clicked_on_ad_dummies = pd.get_dummies(df_encode['Clicked on Ad'], prefix='Clicked_on_Ad', prefix_sep='_')

#Conversion of encode '0' and '1'
gender_dummies = gender_dummies.astype(int)
city_dummies = city_dummies.astype(int)
province_dummies = province_dummies.astype(int)
category_dummies = category_dummies.astype(int)
clicked_on_ad_dummies = clicked_on_ad_dummies.astype(int)

#Concat data
df_encoded = pd.concat([df_encode.drop(columns=['Gender', 'city', 'province', 'category', 'Clicked on Ad']),
                        gender_dummies, city_dummies, province_dummies, category_dummies, clicked_on_ad_dummies], axis=1)

print(df_encoded)
```

# **TASK 3: DATA MODELING**

## **Data Training**

```python
from sklearn.model_selection import train_test_split
```

```python
df_training = df_encoded.copy()
```

```python
df_training.drop(columns=['Timestamp'], inplace=True)
```

```python
#Separating target and feature
x = df_training.drop(columns=['Clicked_on_Ad_No', 'Clicked_on_Ad_Yes'])
y = df_training['Clicked_on_Ad_Yes']

#Dividing data test and train
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print("Data Train:")
print(x_train)
print(y_train)

print("\nData Test:")
print(x_test)
print(y_test)
```

# **First Experiment: Without Normalization**

```python
df_model = df_training.copy()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
#Training model
model = DecisionTreeClassifier(random_state=42)
model.fit(x_train, y_train)

#predict and evaluate model
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print(f'Akurasi tanpa normalisasi: {accuracy}')
print('Confusion Matrix tanpa normalisasi:')
print(cm)
```

```python
#Decision Tree Visualization
from sklearn.tree import plot_tree
plt.figure(figsize=(20,10))
plot_tree(model, feature_names=x_train.columns, class_names=['class0', 'class1'], filled=True)
plt.show()
```

```python
#Visualization Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Clicked_on_Ad_No', 'Clicked_on_Ad_Yes'],
            yticklabels=['Clicked_on_Ad_No', 'Clicked_on_Ad_Yes'])
plt.title('Confusion Matrix Without Normalization')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

## **Second Experiment: With Normlization**

```python
from sklearn.preprocessing import StandardScaler
```

```python
#Normalization Feature
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
#traning model
model = DecisionTreeClassifier(random_state=42)
model.fit(x_train_scaled, y_train)

#predict and evaluate
y_pred = model.predict(x_test_scaled)
accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print(f'Akurasi dengan normalisasi: {accuracy}')
print('Confusion Matrix dengan normalisasi:')
print(cm)
```

```python
#Visualization Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Clicked_on_Ad_No', 'Clicked_on_Ad_Yes'],
            yticklabels=['Clicked_on_Ad_No', 'Clicked_on_Ad_Yes'])
plt.title('Confusion Matrix With Normalization')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

## **Business Recommendation**

So that the previously predicted 5% increase can be achieved, here are several business recommendations proposed:

1. Class 0 (Not Click on Ads)

   This group is likely to be ad-resistant due to high internet exposure. To capture their attention, focus on:

   - **Relevance and Personalization**: Deliver highly targeted ads based on their interests, browsing behavior, and demographics. Ensure ads are directly relevant to their needs and desires.
   - **Incentives and Rewards**: Offer compelling incentives like discounts, exclusive content, or sweepstakes to encourage clicks.
   - **Ad Format and Placement**: Experiment with less intrusive ad formats like native ads or sponsored content. Place ads strategically to avoid ad-blocking software.
   - **Ad Quality**: Invest in high-quality, visually appealing ads that provide value and information.
   - **Trust and Authority**: Build trust through endorsements, certifications, or customer testimonials.
   - **Mobile Optimization**: Ensure ads are optimized for mobile devices as this group likely spends a significant amount of time on smartphones.




2. Class 1 (Click on Ads)

   This group is more likely to click on ads but might require different approaches to increase click-through rates:
   
   - **Clear and Simple Ad**: Use concise and easy-to-understand language that highlights the ad's benefits.
   - **Trust and Authority**: Build trust through endorsements, certifications, or customer testimonials.
   - **Ad Placement**: Consider placing ads in high-visibility areas of the website or app.
